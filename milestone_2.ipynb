```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import collections
```

```python
df = pd.read_csv('Phishing.csv')
```

Count different values in column "Result" and create dataframe for plot

```python
result_counter = collections.Counter(df['Result'])
```

```python
classes = pd.DataFrame.from_records(result_counter.most_common(), columns=['Class','Num_Observations'])
```

Plot bar chart

```python
plt.style.use('ggplot')
barlist = plt.barh(classes['Class'],classes['Num_Observations'])
plt.ylabel('Class')
plt.xlabel('Number of Observations')
plt.title('Class distribution of the websites')
plt.yticks(classes['Class'])
for i, v in enumerate(classes['Num_Observations']):
    plt.text(v, classes['Class'].iloc[i], " "+str(v), color='black', va='center', fontweight='bold')
barlist[0].set_color('b')
barlist[1].set_color('r')
plt.show()
```
